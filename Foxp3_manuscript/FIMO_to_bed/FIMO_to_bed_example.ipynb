```python
#Here we change it to bed format.
def Onestep(df):
    dfA = df.copy()
    dfA=dfA.rename(columns={"sequence_name": "Chr"})
    dfA["TnGName"]=dfA['Chr'] + "_" + dfA["start"].astype(str) + "to" + dfA["stop"].astype(str)
    dfbed=dfA[['Chr', 'start', 'stop','TnGName','score', 'strand']].copy()
    s = dfbed['stop'] < dfbed['start']
    dfbed.loc[s, ['stop','start']] = dfbed.loc[s, ['start','stop']].values
    dfbed=dfbed.dropna()
    dfbed['start']=dfbed['start'].astype('int')
    dfbed['stop']=dfbed['stop'].astype('int')
    return dfbed
```

```python
Human=Onestep(dfHuman)
Mouse=Onestep(dfMouse)
Zebrafish=Onestep(dfZebrafish)
```
